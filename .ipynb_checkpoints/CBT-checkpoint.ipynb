```python


# Load the YOLO model
model_path = os.path.join('.', 'runs', 'detect', 'train', 'weights', 'best.pt')
model = YOLO(model_path)

# Load the image
image_path = "C:/Users/Daithoulung Rongmai/KRS(ML)/CC/Chess-Corner-Detection-1/train/images/IMG_20230913_195331_jpg.rf.244026d2812d546e381b3c61c80cd99a.jpg"
img = Image.open(image_path)
img_np = np.array(img)  # Convert to NumPy array for OpenCV

# Run inference
results = model.predict(img)

center_coordinates = []

custom = [(-30, -30), (30, 30), (30, -30), (-30, 30)]
# Extract the bounding box coordinates (corners) from results
for result in results:
    # result.boxes.xyxy gives you the coordinates of the bounding boxes
    boxes = result.boxes.xyxy.cpu().numpy()  # Get the bounding box coordinates in (x_min, y_min, x_max, y_max)

    for i in range ( len(boxes)):
        x_min, y_min, x_max, y_max = boxes[i]

        center_x = int((x_min + x_max) / 2)+custom[i][0]
        center_y = int((y_min + y_max) / 2)+custom[i][1]

        center_coordinates.append([center_x, center_y])

        img_with_center = img_np.copy()

        cv2.circle(img_with_center, (center_x, center_y), 5, (0, 255, 0), -1)  # Green circle for the center point+
        # Display the image with the center point
        plt.imshow(cv2.cvtColor(img_with_center, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB for correct display in Matplotlib
        plt.axis('off')
        plt.show()
```

```python
center_coordinates_np = np.array(center_coordinates)

# Ensure we have exactly 4 points to form a box
if len(center_coordinates_np) == 4:
    sorted_centers = center_coordinates_np[center_coordinates_np[:, 1].argsort()]
    top_left, top_right = sorted(sorted_centers[:2], key=lambda x: x[0])
    bottom_left, bottom_right = sorted(sorted_centers[2:], key=lambda x: x[0])
    cropped_image = img_np[int(y_min):int(y_max), int(x_min):int(x_max)]
    # Source points (the detected corners of the chessboard)
    src_points = np.array([top_left, top_right, bottom_right, bottom_left], dtype="float32")

    # Destination points (mapping to a perfect 400x400 square)
    grid_size = 600  # We will map it to a 400x400 square
    dst_points = np.array([[0, 0], [grid_size - 1, 0], [grid_size - 1, grid_size - 1], [0, grid_size - 1]], dtype="float32")

    # Compute the perspective transform matrix
    M = cv2.getPerspectiveTransform(src_points, dst_points)

    # Apply the perspective transformation to the image
    warped_image = cv2.warpPerspective(img_np, M, (grid_size, grid_size))

    # Divide the warped image into an 8x8 layout
    rows = 8
    cols = 8

    # Calculate the width and height of each cell
    cell_width = grid_size // cols
    cell_height = grid_size // rows

    # Draw the grid on the warped image
    img_with_grid = warped_image.copy()

    # Draw horizontal lines
    for i in range(1, rows):
        y = i * cell_height
        cv2.line(img_with_grid, (0, y), (grid_size, y), (255, 0, 0), 1)  # Horizontal line

    # Draw vertical lines
    for i in range(1, cols):
        x = i * cell_width
        cv2.line(img_with_grid, (x, 0), (x, grid_size), (255, 0, 0), 1)  # Vertical line

    # Display the warped image with the grid
    plt.imshow(cv2.cvtColor(img_with_grid, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB for correct display in Matplotlib
    # plt.axis('off')
    plt.show()
else:
    print(f"Expected 4 center points but got {len(center_coordinates_np)}")
```
